```python
# Split and show train
ts = ts6['item 38760'].to_frame()

ts = ts.dropna()
```

```python
# Calculate KPSS test
def kpss_test(data):
    kpss_test_result = stattools.kpss(np.array(data))
    print('\nThe KPSS test statistic: %f' % kpss_test_result[0])
    print('The p-value of the test: %f' % kpss_test_result[1])
    
# Calculate The Augmented Dickey-Fuller test
def adfuller_test(data):
    test_result = stattools.adfuller(np.array(data))
    print('\nThe Augmented Dickey-Fuller test statistic: %f' % test_result[0])
    print('The p-value of the test: %f'% test_result[1])
```

```python
from scipy.stats import boxcox

print('Original')
kpss_test(ts['item 38760'])
adfuller_test(ts['item 38760'])
ts['item 38760'].plot(figsize=(17, 3))

# If lmbda is None, find the lambda that maximizes the log-likelihood function and return it as the second output argument.
ts['stat'] = boxcox(ts['item 38760'] + 0.01)[0]

print('\nAfter Box-Cox')
kpss_test(ts['stat'])
adfuller_test(ts['stat'])

ts['stat'].plot(figsize=(17, 3), color="green")
```

p-value <= 0.05: Reject the null hypothesis (H0), the data does not have a unit root and is stationary.

```python
from statsmodels.tsa.stattools import acf, pacf, adfuller
from statsmodels.tsa.arima_model import ARIMA
import matplotlib.pyplot as plt

def show_acf_and_pacf(data):
    acf_ = acf(data)
    pacf_ = pacf(data)

    print(adfuller(data)[1])
    plt.figure(figsize=(15, 5))
    plt.title('ACF')
    plt.stem(acf_)
    plt.show()

    plt.figure(figsize=(15, 5))
    plt.title('PACF')
    plt.stem(pacf_)
    plt.show()
```

```python
print('Original')
show_acf_and_pacf(ts['item 38760'])

print('After Box-Cox')
show_acf_and_pacf(ts['stat'])
```

```python
# Split into train and test

train_size = len(ts) * 8 / 10
train = ts['stat'].head(train_size)
test = ts['stat'].tail(int(len(ts) - train_size)).to_frame()
```

```python
import statsmodels.api as statmodels

p_best = 0
q_best = 0
aic_best = 9000
for p in range(4):
    for q in range(4):
        model = statmodels.tsa.ARMA(train, (p, q)).fit()
        if (model.aic < aic_best):
            p_best = p
            q_best = q
            aic_best = model.aic
            
print('Best aic: %f', aic_best)
print('Best p: %f', p_best)
print('Best q: %f', q_best)
```

```python
# Check of Residual Characteristics
arma = statmodels.tsa.ARMA(train, (p_best, q_best)).fit()
residual = arma.resid.values
kpss_test(residual)
adfuller_test(residual)
show_acf_and_pacf(residual)
```

```python
from scipy.stats import probplot, wilcoxon

print(wilcoxon(model.resid.values))

x = probplot(model.resid, dist="norm", plot=plt)
```

```python
plt.figure(figsize=(15,5))
(train).plot()
arma.predict().plot()
plt.show()
```

```python
def SMAPE(x, y):
    sum_ = np.abs(x - y) / (np.abs(x) + np.abs(y))
    return np.mean(sum_)
```

```python
forecast = arma.forecast(steps=len(test))[0]
print(SMAPE(forecast, test['stat']))
```

**II. Find the best parameters of ARIMA automaticaly** 
* 0) read data for TS under consideration and use the first 80% of all length of the TS as train TS.  
* 1) plot the graph of the train TS, preprocess data (delete tail, make calendar transformation if needed)
* 2) stabilize dispersion of train TS through box-cox transformation if needed.
* 3) Run grid search for ARIMA parameters (use transformed train TS) according to AIC. Use at least 10 distinct values for $p,q,d$. How many time does it take?
* 4) Build forecast for the initial TS with best ARIMA model selected automaticaly at step 2 (you need to transform it like in step 2).
* 5) Build forecast interval for the last 20% of transformed TS. Calculate loss MACAPE of your forecast. Does the forecast look appropriate?

```python
import statsmodels.api as statmodels

p_best = 0
d_best = 0
q_best = 0
aic_best = 9000
for p in range(9):
    for d in range(1):
        for q in range(9):
            try:
                model = statmodels.tsa.ARIMA(train, (p, d, q)).fit()
                if (model.aic < aic_best):
                    p_best = p
                    d_best = d
                    q_best = q
                    aic_best = model.aic
            except:
                continue
            
print('Best aic: %f', aic_best)
print('Best p: %f', p_best)
print('Best d: %f', d_best)
print('Best q: %f', q_best)
```

```python
arima = statmodels.tsa.ARIMA(train, (p_best, d_best, q_best)).fit()
forecast = arima.forecast(steps=len(test))[0]
print(SMAPE(forecast, test['stat']))
```
